```python
def categorize_sentiment(score):
    if score >= 4:
        return 'positive'
    elif score == 3:
        return 'neutral'
    elif score <= 2:
        return 'negative'
    else:
        return 'unknown'

df_original['categorized_sentiment'] = df_original['score'].apply(categorize_sentiment)

# Display the head of the DataFrame with the new sentiment column
(df_original[['content', 'score', 'categorized_sentiment']].head())
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
df = df.sort_index()
df_original = df_original.sort_index()
```

```python
combined_df = pd.DataFrame({
    'transformer_sentiment': df['sentiment'],
    'rating_sentiment': df_original['categorized_sentiment']
})
```

```python
print(combined_df.head())

# Calculate the confusion matrix
cm = confusion_matrix(combined_df['rating_sentiment'], combined_df['transformer_sentiment'], labels=['positive', 'neutral', 'negative'])

# Display the confusion matrix
print('Confusion Matrix:')
cm
```

```python
print('Classification Report:')
print(classification_report(combined_df['rating_sentiment'], combined_df['transformer_sentiment'], labels=['positive', 'neutral', 'negative']))
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['positive', 'neutral', 'negative'], yticklabels=['positive', 'neutral', 'negative'])
plt.xlabel('Predicted Sentiment')
plt.ylabel('Actual Sentiment')
plt.title('Confusion Matrix Heatmap')
plt.show()
```
